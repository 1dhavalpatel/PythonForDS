## importing necessary modules

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```python
stopwords = set(STOPWORDS)
```

### Set the content for the word cloud

```python
# Set the list of words - This could be loaded from csv or by some other way as well

comments = ['future technology',
            'There are three ways to make money from decentralized lending: investing in the individual governance tokens, investing in the Ethereum blockchain, or yield farming with stablecoins',
            'Decentralized Finance is the future, go DeFi',
            'The simplest way to earn a passive income through DeFi is to deposit your cryptocurrency onto a platform or protocol that will pay you an APY (annual percentage yield) for it.',
            'Highly programmable smart contracts automate execution and enable the creation of new financial instruments and digital assets.',
            'everything is a Token in DeFi',
            'Short for decentralized finance, DeFi is an umbrella term for peer-to-peer financial services on public blockchains, primarily Ethereum.',
            'The Ethereum blockchain ecosystem gave the world new technological capabilities, including digital assets, smart contracts, and decentralized applications (DApps)',
            'Decentralized finance (DeFi) is an emerging financial technology based on secure distributed ledgers similar to those used by cryptocurrencies.',
            'The goal of DeFi is to get rid of the third parties that are involved in all financial transactions',
            'Decentralized finance (DeFi). A global, open alternative to the current financial system. Products that let you borrow, save, invest, trade, and more.',
            'Decentralized finance offers financial instruments without relying on intermediaries such as brokerages, exchanges, or banks by using smart contracts on a blockchain.'
            ]
```

```python
text = [word for word in comments if word not in stopwords]
text=" ".join(text)
text
```

### Build the Word Cloud

```python
wordcloud = WordCloud(stopwords=stopwords, 
                      background_color="white").generate(text)
# Display the generated Word Cloud
plt.imshow(wordcloud, interpolation='bilinear')   #'interpolation=bilinear' is used to make the image appear smoother.
plt.axis("off")
plt.show()
```

## Some WordCloud customizations 

By default, the word cloud uses bi-grams (pairs of words) instead of single words. 
<br>If needed, we can turn this off when we instantiate the WordCloud object by changing the parameter collocations=False
<br>max_words=50 --> to restrict the number of words in the cloud
<br>colormap='tab20c' --> use the colormap of your choice

```python
wordcloud2 = WordCloud(width = 1000, height = 800,
                       max_words=50,
                       colormap='tab20c',
                       collocations = False).generate(text)


plt.imshow(wordcloud2)
plt.axis("off")
plt.show()
```

### Changing the shape of a word cloud

```python
import numpy as np
from PIL import Image
```

```python
# Create an array from the image you want to use as a mask
heart_mask = np.array(Image.open('Heart_B.png'))

#Tip - The mask should have very distinctive color & shape, and the background should be white
```

```python
wordcloud3 = WordCloud(scale=3,
                       max_words=50,
                       #colormap='RdYlGn',
                       colormap='tab20c',
                       mask=heart_mask,
                       background_color='white',
                       stopwords=stopwords,
                       collocations=False).generate(text)

plt.figure(figsize=(10,8))
plt.imshow(wordcloud3)
plt.axis('off')
plt.title('Heart Shape Word Cloud')
plt.show()
```

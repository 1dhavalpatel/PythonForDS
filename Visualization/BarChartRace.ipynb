## Importing libraries

```python
import pandas as pd
import bar_chart_race as bcr
```

## Loading and preprocessing data

```python
df_temperature = pd.read_csv('Monthly_Mean_Temp_IMD-2901_to_2019_0.csv')
```

```python
df_temperature.head()
```

```python
df_temperature.columns
```

```python
# Drop unwanted columns

df_temperature = df_temperature.drop(['ANNUAL', 'JAN-FEB', 'MAR-MAY', 'JUN-SEP','OCT-DEC'], axis = 1)
df_temperature
```

```python
# Set index on year column --> Very imp, without setting index on frequency column
# bar_chart_race will not work

df_temperature = df_temperature.set_index('YEAR')
df_temperature
```

```python
df_temperature.info()
```

```python
df_temperature.isnull().any()
```

## Creating Bar Chart Race

### creating a basic chart

```python
bcr.bar_chart_race(df = df_temperature, title = "India - Monthly Mean Temp IMD")
```

### Customizing the chart

Explore more options here --> https://www.dexplo.org/bar_chart_race/tutorial/

```python
bcr.bar_chart_race(df = df_temperature, 
                   title = "India - Monthly Mean Temp IMD: 1901 to 2019",
                   orientation = 'h' , # 'h' , 'v'
                   fixed_order = True, # default: False
                   fixed_max=True, # default: False
                   #n_bars=6, # limiting number of bars
                   #label_bars=False, # default: True
                   #sort='asc', # default: 'desc' 
                   steps_per_period = 2, # default: 10
                   period_length = 1000, # default: 500 (half a sec),
                   figsize=(5, 3), # default: (6, 3.5)
                   dpi=100, # default: 144
                   #cmap='OrRd',
                   period_fmt='Year - {x:.0f}'
                  )
```

```python
import pandas as pd
import numpy as np
```

```python
data = pd.DataFrame(np.random.rand(10, 4), columns=['A','B','C','D'])
```

```python
data
```

## Basic Plots

```python
data.plot.bar();
```

```python
data.plot.bar(stacked=True);
```

```python
data.plot.barh(stacked=True);
```

```python
data.plot.area(stacked=False, alpha = 0.4) #alpha (transparency) is set to 0.5 by default
```

```python
data.plot.box()
```

## Subplots

```python
data = pd.DataFrame(np.random.rand(10, 4), columns=['A','B','C','D'])
```

```python
data.plot(subplots=True,figsize=(20,10))
```

```python
data.plot.bar(subplots=True,layout=(2, 2),figsize=(20,10))
```

```python
import pandas as pd
import yfinance as yf

import datetime
from datetime import date, timedelta

cryptoSymbol = 'ETH-USD'
```

```python
today = date.today()

d1 = today.strftime("%Y-%m-%d")
end_date = d1

d2 = date.today() - timedelta(days=730)
d2 = d2.strftime("%Y-%m-%d")
start_date = d2
```

```python
data = yf.download(cryptoSymbol, 
                      start=start_date, 
                      end=end_date, 
                      progress=False)
```

```python
data.head()
```

```python
data["Date"] = data.index
data = data[["Date", "Open", "High", "Low", "Close", "Adj Close", "Volume"]]
data.reset_index(drop=True, inplace=True)
```

```python
data.tail()
```

```python
data.shape
```

```python
import plotly.graph_objects as go

figure = go.Figure(data=[go.Candlestick(x=data["Date"],
                                        open=data["Open"], 
                                        high=data["High"],
                                        low=data["Low"], 
                                        close=data["Close"])])

figure.update_layout(title = cryptoSymbol + " Price Analysis", 
                     xaxis_rangeslider_visible=False)

figure.show()
```

```python
correlation = data.corr()
print(correlation["Close"].sort_values(ascending=False))
```

## Cryptocurrency Price Prediction Model

```python
from autots import AutoTS

model = AutoTS(forecast_length=15, frequency='infer', ensemble='simple')
model = model.fit(data, date_col='Date', value_col='Close', id_col=None)
```

```python
prediction = model.predict()
forecast = prediction.forecast
print(forecast)
```

```python
forecast.index
```

```python
model_results = model.results()
validation = model.results("validation")
```

### Compare Show votings between IMDB and TMDB over the years

```python
shows = df_titles[df_titles.type == 'SHOW'][['release_year', 'imdb_score', 'tmdb_score']]
shows
```

```python
shows.dropna(axis=0, subset=['imdb_score', 'tmdb_score'], inplace=True)
shows
```

```python
# Group by year
grouped_df = shows.groupby(["release_year"]).mean()
grouped_df
```

```python
plt.plot(grouped_df.index, grouped_df.imdb_score, label="IMDB Score Avg", linewidth=2, marker='o', markersize=7)
plt.plot(grouped_df.index, grouped_df.tmdb_score, label="TMDB Score Avg", linewidth=2, marker='o', markersize=7, linestyle="--")
plt.xlabel("Year")
plt.ylabel("Average")
plt.legend()
plt.show()
```

After 1990 TMDB Avg Score seems to be consistently higher than IMDB Avg Score

### Create a Facet Grid for movies dataset visualizing tmdb score by genre by year - top  3 years

```python
years = df_titles[df_titles.type == 'MOVIE'].release_year.value_counts().head(3)
years
```

```python
df_titles_filtered = df_titles[(df_titles.type == 'MOVIE') & (df_titles.release_year.isin(years.index.tolist()))]
df_titles_filtered
```

```python
# Since genre consists of a list of items, considering the first value from the list for this analysis
### This is just one of the ways to get a value from the list, there are other strategies as well to get all the values

import ast

def get_first_value_categories(arr):
    arr = ast.literal_eval(arr)
    
    if len(arr) == 0:
        return np.nan
    
    else:
        return arr[0]
    

df_titles_filtered['genres'] = df_titles_filtered['genres'].apply(get_first_value_categories)
```

```python
genre_list = df_titles_filtered[df_titles_filtered.type == 'MOVIE'].genres.value_counts().head(3)
genre_list
```

```python
df_titles_filtered = df_titles_filtered[df_titles_filtered.genres.isin(genre_list.index.tolist())]
df_titles_filtered
```

```python
df_titles_filtered.dropna(axis=0, subset=['imdb_score', 'tmdb_score'], inplace=True)
df_titles_filtered
```

```python
plt.rcParams['figure.figsize'] = 10,5

fg = sns.FacetGrid(df_titles_filtered, row='genres', col='release_year', hue='genres')
fg = fg.map(plt.scatter, 'imdb_score', 'tmdb_score')

plt.show()
```

```python
df_titles_filtered = df_titles_filtered[['id','genres','tmdb_score']]
df_titles_filtered
```

```python
mean_values = [df_titles_filtered[df_titles_filtered.genres == 'comedy'].mean(),
               df_titles_filtered[df_titles_filtered.genres == 'drama'].mean(),
               df_titles_filtered[df_titles_filtered.genres == 'documentation'].mean()]
mean_values
```

```python
score_plot = sns.barplot(data=mean_values)

score_plot.set_xlabel('Genres')
score_plot.set_xticklabels(['Comedy', 'Drama', 'Documentation'])
score_plot.set_ylabel('Average TMDB Score')
score_plot.set_title('Average TMDB Score by Genre')

plt.show()
```
